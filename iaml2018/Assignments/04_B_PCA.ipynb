## Description of the dataset

In this part of the assignment we will explore the MNIST digits dataset. We expect the digits to lie in a lower-dimensional manifold and want to examine the representation we get by applying Principal Components Analysis (PCA). PCA maps the data into a new space by effectively rotating the base vectors of the input space to the directions with the highest variance.

### Description of the Data

This MNIST Dataset is a collection of handwritten digits. The samples are partitioned (nearly) evenly across the 10 different digit classes {0, 1, . . . , 9}. We use a preprocessed version for which the data are $8 \times 8$ pixel images containing one digit each. For further details on how the digits are preprocessed, see the sklearn documentation. The images are grayscale, with each pixel taking values in {0, 1, . . . , 16}, where 0 corresponds to black (weakest intensity) and 16 corresponds to white (strongest intensity). Therefore, the dataset is a N × 64
dimensional matrix where each dimension corresponds to a pixel from the image and N is the number of
images.

Set `svd_solver='full'` and keep all other parameters at default values. Then plot the *percentage* variance explained by each dimension as a `bar`-graph, in descending order of variance. *Hint: see the `explained_variance_ratio_` attribute of the PCA object.* On the same graph, plot (line) the cumululative explained variance as you add components (again in descending order of variance explained). Indicate the point at which 80% of the variance is explained (e.g. using matplotlib's [`vlines()`](https://matplotlib.org/2.2.3/api/_as_gen/matplotlib.pyplot.vlines.html)): label the number of components at this point (e.g. in the Legend). Make sure to distinguish between the plots and label them appropriately.<br>&nbsp;&nbsp;*Tip: you may need to use [`twinx()`](https://matplotlib.org/2.2.3/gallery/api/two_scales.html#sphx-glr-gallery-api-two-scales-py) to better combine the two plots. &nbsp;&nbsp;**(b)** [Text] By commenting on the shape of the plot, why do you think 80% of the variance seems like a reasonable threshold? *Hint: Think about why we do PCA*. (<3 sentences)<br>
&nbsp;&nbsp;**(c)** [Text] What can you say about the number of components required to explain 80% of the variance? (<4 sentences)

````python
#(a) # Your Code goes here:

# Train the full dimentional PCA and output the explained_variance _ratio
pca = PCA(n_components=data.shape[1], svd_solver='full')
pca.fit(data)
evr = pca.explained_variance_ratio_

# barplot the explained_variance_ratio_
sns.set(style="white")
fig, ax1 = plt.subplots(figsize=(18,10))
ax1 = sns.barplot(x=np.arange(evr.shape[0]), y=evr) 
ax1.set_xlabel('n_components')
ax1.set_ylabel('Explained-Variance-Ratio')

# compute and plot the cumulative explained_variance_ratio_ 
cumulative_evr = [ sum(evr[0:i+1]) for i in range(evr.shape[0]) ] 
ax2 = ax1.twinx() # instantiate a second axes that shares the same x-axis
sns.lineplot(data=pd.DataFrame(cumulative_evr), color="b", ax=ax2, legend=False)
ax2.set_ylabel('Cumulative Explained-Variance-Ratio')  # we already handled the x-label with ax1

# Find the number of dimensions that 80% of variance is explained
# Here I use the trick that numpy.argmax return the index of the first true value
eighty_explained_point = np.argmax(np.array(cumulative_evr) >= 0.8)
plt.vlines(eighty_explained_point, 0, 1, linestyle="dashed")
plt.annotate("80% variance explained", xy=(eighty_explained_point - 4, 1)) # label the number of components at this point 

plt.show()
````

(b) ***Your answer goes here:***

At the point of 80% variance explained, we have a **significant decrease of dimensions** and it looks like an **"elbow"** which represents where we start to have diminishing returns by increasing n_components. Hence, 80% is a reasonable ratio.

(c) ***Your answer goes here:***

The number of the components required to explain 80% of the variance is **at least 13** (0-12 from the above graph). These components (which are orthgonoal to each other) preserve a large portion of information while significantly reduce the number of dimensions. So in fact, there are **many redundant dimensions**.

````python
#(b) # Your Code goes here:
nums = [0, 1, 2, 5, 15, 30]
plt.subplots(figsize=(15, 15))
i = 1 # position index
for n in nums:
    reconstruct_digit = pca.mean_ + np.dot(trans[0: n], pca.components_[0: n]) # use only n components for reconstruction
    ax = plt.subplot(3, 2, i)
    sns.heatmap(reconstruct_digit.reshape(8, 8), vmin=-1.5, vmax=14.5,linewidths=.01, cmap='binary', ax=ax)
    ax.set_title("[{} components] MAD: {:.3f}".format(n, np.mean(np.abs(reconstruct_digit - ori_digit))))
    i += 1
plt.show()
````
